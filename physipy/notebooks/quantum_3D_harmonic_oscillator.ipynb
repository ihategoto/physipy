```python
import numpy as np
import physipy as phys
```

```python
E = np.arange(1.4, 5, 0.1)
l = np. arange(0, 3, 1)
potential = phys.harmonic
psi_0_outward = 0
grid = phys.Grid(1e-4, 10, 1e-3)
solver = phys.SolverOpts()
```

```python
eigenstates = phys.energy_levels(E, l, potential, psi_0_outward, grid = grid, solver = solver)
```

```python
import matplotlib.pyplot as plt

for eigenstate in eigenstates:
    plt.title("Energy = {:.3f}, l = {}".format(eigenstate.E, eigenstate.l))
    plt.plot(eigenstate.coord, eigenstate.psi)
    plt.show()
```
