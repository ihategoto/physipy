```python
import numpy as np
```

```python
def neuron(x, w):
    return np.sign(np.sum(x*w.T, axis = 1))

def backpropagate(y, y_pred, x, w, alpha = .01):
    y_backpropagate = np.copy(y)
    y_backpropagate[y == y_pred] = 0
    temp = alpha * np.sum(y_backpropagate[:, np.newaxis] * x, axis = 0)
    print(f'adding {temp}')
    w += temp
    return w
```

```python
rng = np.random.default_rng(seed = 13)
training_set = rng.random((700, 3))
test_set = rng.random((300, 3))
w = np.array([-.545, .2])#rng.random(2)
training_set[:, 2] =  np.where(training_set[:,0] > training_set[:,1], 1, -1)
test_set[:, 2] = np.where(test_set[:,0] > test_set[:,1], 1, -1)

epochs = 100

for epoch in range(epochs):
    y_pred = neuron(training_set[:, :2], w)
    results = np.where(y_pred == training_set[:, 2], 1, 0)
    print(f'Epoch = {epoch} Weights = {w} Correct = {np.sum(results)}')
    w = backpropagate(training_set[:, 2], y_pred, training_set[:, :2], w)
```
